## 2. Chargement des données

```python
# Charger votre dataset ici
df_disponibilite = pd.read_csv('..\\Data\\Clean\\DF_DisponibiliteAlimentaire_2017.csv')
df_distance = pd.read_csv('..\\Data\\Clean\\DF_Distances.csv')
df_pib = pd.read_csv('..\\Data\\Clean\\DF_Pib_2017.csv')
df_stabilite = pd.read_csv('..\\Data\\Clean\\DF_Political_Stability_2017.csv')
df_population = pd.read_csv('..\\Data\\Clean\\DF_Population_2017.csv')
df_nourriture = pd.read_csv('..\\Data\\Clean\\DF_Nourriture_Quali_2017.csv')
df_business = pd.read_csv('..\\Data\\Clean\\DF_Business.csv')
df_Nourriture_Quali = pd.read_csv('..\\Data\\Clean\\DF_Nourriture_Quali_2017.csv')
```

```python
df_disponibilite
```

```python
df_distance.head()
```

```python
df_merged = pd.merge(df_disponibilite, df_distance[['Country Code', 'Distance en KM']], on='Country Code', how="left")
```

```python
df_merged
```

```python
df_pib.head()
```

```python
df_merged2 = pd.merge(df_merged,df_pib [['PIB par habitant','Country Code']], on='Country Code', how='left') 
```

```python
df_merged2.head()
```

```python
df_stabilite.head()
```

```python
df_merged3 = pd.merge(df_merged2,df_stabilite [['Indice de stabilité politique','Country Code']], on='Country Code', how='left').reset_index(drop=True) 
```

```python
df_merged3 = df_merged3.dropna(subset=["Country Name"])
```

```python
df_merged3.head()
```

```python
# # voir si merge se fait sur Country Code ou left_on'Country Code', right_on='iso3'
df_merged4 = pd.merge(df_merged3,df_Nourriture_Quali [['Population pouvant acheter une alimentation saine (%)','Country Code']], on='Country Code', how='left') 
```

```python
df_merged5 = pd.merge(df_merged4,df_population [['Population','iso3']], left_on='Country Code',right_on='iso3', how='left').reset_index(drop=True) 
```

```python
df_Clean = pd.merge(df_merged5,df_business [['Taux de facilité de faire des affaires (%)','Country Code']], on='Country Code', how='left').reset_index(drop=True) 
```

```python
df_Clean
```

```python
df_Clean.drop(columns=['Unnamed: 0','Zone','nom_fr','country_en','iso2','iso3'], inplace=True)
```

```python
df_Clean
```

```python
df_Clean.insert(0, 'Zone', df_Clean.pop('Country Name'))
df_Clean.insert(1, 'Country Code', df_Clean.pop('Country Code'))
```

```python
df_Clean.to_csv('..\\Data\\Clean\\DF_Clean.csv', index=False)
```

### Vérification de la validité des données

```python
# Classement des pays par ordre décroissant de population
df_Clean = df_Clean.sort_values(by='Population', ascending=False)

# Calcul du cumul de la population
df_Clean['Cumul Population'] = df_Clean['Population'].cumsum()
```
